## Why we use XGBoost (short summary)

- Combines many small decision trees to make strong predictions  
- Each tree corrects errors from previous ones (boosting)  
- Handles nonlinear relationships and feature interactions well  
- Built-in regularization helps prevent overfitting  
- Works extremely well on structured/tabular data  
- Often gives high accuracy with relatively little feature engineering
